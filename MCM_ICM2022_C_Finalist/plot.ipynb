```python
import plotly.graph_objects as go
animals=['Linear Regression', 'LR-LSTM model']
y1, y2 = [7.724, 8.480], [13.116, 18.736]
fig = go.Figure(data=[
    go.Bar(name='MAE', x=animals, y=y1, marker_color='indianred', text=y1, textposition="outside"),
    go.Bar(name='RMSE', x=animals, y=y2, marker_color='lightsalmon', text=y2, textposition="outside")])
# Change the bar mode
fig.update_layout(title={'text':'Errors on Gold data','y' : 0.9,'x' : 0.5,  }, height=600, width=500, legend_title_text='Type of error')
fig.show()
```
